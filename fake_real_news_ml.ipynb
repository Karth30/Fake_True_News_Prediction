```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df_final["text"] = df_final["text"].apply(wordopt)
```

```python
x = df_final["text"]
y = df_final["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
import pickle
import numpy as np
import pandas as pd
import os  # To handle directory creation

def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt) 
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)

    # Predictions from all models
    prob_lr   = LR.predict_proba(new_xv_test)[:, 1]  # Probability of "Not A Fake News" (class 1)
    prob_dt   = DT.predict_proba(new_xv_test)[:, 1]
    prob_gb   = GB.predict_proba(new_xv_test)[:, 1]
    prob_rc   = RC.predict_proba(new_xv_test)[:, 1]
    prob_svm  = SVM.predict_proba(new_xv_test)[:, 1]

    # Find the maximum probability and corresponding model
    all_probs = np.array([prob_lr[0], prob_dt[0], prob_gb[0], prob_rc[0], prob_svm[0]])
    max_prob_index = np.argmax(all_probs)
    
    # Choose the best model prediction based on max probability
    model_predictions = [prob_lr, prob_dt, prob_gb, prob_rc, prob_svm]
    best_prediction = model_predictions[max_prob_index]
    
    # Assign the final prediction based on the best model's prediction (threshold > 0.5 for "Not A Fake News")
    final_pred = 1 if best_prediction > 0.5 else 0
    
    # Print individual model predictions and final best model prediction
    print("\n\nModel Predictions (Probabilities):")
    print(f"LR Prediction:   {output_label(prob_lr[0])} - Probability: {prob_lr[0]:.4f}")
    print(f"DT Prediction:   {output_label(prob_dt[0])} - Probability: {prob_dt[0]:.4f}")
    print(f"GBC Prediction:  {output_label(prob_gb[0])} - Probability: {prob_gb[0]:.4f}")
    print(f"RFC Prediction:  {output_label(prob_rc[0])} - Probability: {prob_rc[0]:.4f}")
    print(f"SVM Prediction:  {output_label(prob_svm[0])} - Probability: {prob_svm[0]:.4f}")

    print(f"\nFinal Best Model Prediction (based on max probability): {output_label(final_pred)}")

    # Directory where the pickle file will be saved
    pickle_file_path = r"C:\Projects\Fake_Real_News_Prediction\app\best_prediction.pkl"

    # Ensure the directory exists
    os.makedirs(os.path.dirname(pickle_file_path), exist_ok=True)
    
    try:
        # Save the final prediction to a pickle file
        with open(pickle_file_path, "wb") as f:
            pickle.dump(final_pred, f)
        print(f"Pickle file saved at: {pickle_file_path}")
    except Exception as e:
        print(f"An error occurred while saving the pickle file: {e}")
    
    return final_pred
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
import pickle
import os

# Example: assume you already have x_train and y_train defined and preprocessed
# x_train = [...]  # list of training texts
# y_train = [...]  # corresponding labels

# Vectorization
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)

# Train the SVM model
SVM = LinearSVC()
SVM.fit(xv_train, y_train)

# Directory to save model files
model_dir = r"C:\Projects\Fake_Real_News_Prediction\app"
os.makedirs(model_dir, exist_ok=True)

# Save paths
model_path = os.path.join(model_dir, "svm_model.pkl")
vectorizer_path = os.path.join(model_dir, "vectorizer.pkl")

try:
    # Save SVM model
    with open(model_path, "wb") as model_file:
        pickle.dump(SVM, model_file)
    print(f"SVM model saved at: {model_path}")

    # Save vectorizer
    with open(vectorizer_path, "wb") as vec_file:
        pickle.dump(vectorization, vec_file)  # corrected this line to use 'vectorization'
    print(f"Vectorizer saved at: {vectorizer_path}")

except Exception as e:
    print(f"Failed to save model or vectorizer: {e}")
```
